Data Types

Integer: int , Description: Whole numbers such as 3, 1, 200, 300, or 200

Floating Point: float, Description: Numbers with a decimal point: 2.3, 4.6, 200.0

Strings: str, Description: Ordered sequence of characers: "hello", "Sammy", "2000", "New York City" (can be double or single quotes) 

Lists: list, Description: Ordered sequence of objects: [10,"hello",200.3]

Dictionaries: dict, Description: Unordered Key- Value pairs: {"mykey":"value","name":"Frankie"}

Tuples: tup, Description: Ordered immutable sequence of objects: (10,"hello",200.3)

Sets: set, Description: Unpordered collection of unique objects: {"a","b"}

Booleans: bool, Description: Logical value indicating True or False
